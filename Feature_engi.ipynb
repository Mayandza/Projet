```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def create_age(mu=50, sigma=13,mum_samples=100, seed=42):
         np.random.seed(seed)
         sample_age = np.random.normal(loc=mu, scale=sigma, size=mum_samples)
         sample_age = np.round(sample_age, decimals=0)

         return sample_age
```

```python
sample = create_age()
sample
```

```python
def pourcentage_missing(df):
  pourcentage_nan = 100 * df.isnull().sum() / len(df)
  pourcentage_nan = pourcentage_nan[pourcentage_nan > 0].sort_values()

  return pourcentage_nan
```

```python
pourcentage = pourcentage_missing(df)
pourcentage
```

```python
plt.figure(figsize=(14,4), dpi=100)
sns.barplot(x = pourcentage.index, y=pourcentage, palette = 'viridis')
plt.xticks(rotation = 90)
```

```python
pourcentage[pourcentage < 1]
```
