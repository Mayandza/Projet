```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
#creat spark session
spark = SparkSession .builder.appName('spark').getOrCreate()
```

```python
print('Saprk')
```

```python
# Importation des données
df = spark.read.format("csv").options(header = "True").load('football_matches.csv')
```

```python
df.show()
```

```python
## Renommer les colonnes
df = df.selectExpr(
    "*",
    "`FTHG` AS `HomeTeamGoal`",
    "`FTAG` AS `AwayTeamGoal`",
    "`FTR` AS `FinalResult`"

)
```

```python
df.show()
```

```python
## Supprimer les colonnes

df = df.drop("FTHG","FTAG","FTR")
df.limit(10).show()
```

```python
## Afficher le tableau des données comme sur pandas
df.toPandas().head(30)
```

```python
# creation de nouvelle colonne

df = df.withColumn("HomeWin", when(col("FinalResult") == "H", 1).otherwise(0))\
       .withColumn("AwayWin", when(col("FinalResult") == "A", 1).otherwise(0))\
       .withColumn("Null", when(col("FinalResult") == "D", 1).otherwise(0))
df.toPandas().head(10)
```

```python
## creation d'un data frame de bundesliga

df_bun = df.filter((col('Div') == 'D1') &
                  (col('Season') >= 2000 ) &
                  (col('Season') <= 2015 ))

df_bun.toPandas().head(10)
```

```python
## Create math state

df_home = df_bun.groupby('Season','HomeTeam') \
          .agg(sum('HomeWin'). alias('nbHomewin'),
               sum('AwayWin'). alias('nbAwayWin'),
               sum('Null'). alias('nbNull'),
               sum('HomeTeamGoal'). alias('nbHomeTeamGoal'),
               sum('AwayTeamGoal'). alias('nbAwayTeamGoal')) \
          .withColumnRenamed('HomeTeam','Team')
df_home.toPandas().head(10)
```

```python
df_away = df_bun.groupby('Season','AwayTeam') \
          .agg(sum('HomeWin'). alias('nbAHomewin'),
               sum('AwayWin'). alias('nbAAwayWin'),
               sum('Null'). alias('nbANull'),
               sum('HomeTeamGoal'). alias('nbAHomeTeamGoal'),
               sum('AwayTeamGoal'). alias('nbAAwayTeamGoal')) \
          .withColumnRenamed('AwayTeam','Team')
df_away.toPandas().head(10)
```
